```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import plotly.express as px

# Load data
df = pd.read_csv('https://raw.githubusercontent.com/Pravignapala/Principle_of_Data_Science/refs/heads/main/StudentsPerformance.csv')

df.head()

numerical_df = df.select_dtypes(include=['int64', 'float64'])



# Group data by 'parental level of education' and calculate the average math score
parental_level_education_avg_math = df.groupby('parental level of education')['math score'].mean().reset_index()

# Create a bar plot using Plotly Express
fig = px.bar(parental_level_education_avg_math, x='parental level of education', y='math score',
             title='Average Math Score by Parental Level of Education',
             labels={'parental level of education': 'Parental Level of Education', 'math score': 'Average Math Score'})
fig.show()

print("Conclusion: Students whose parents have a master's degree tend to have the highest average math scores, while students whose parents have some high school education tend to have the lowest.")


# Group data by 'race/ethnicity' and calculate the average reading score
race_ethnicity_avg_reading = df.groupby('race/ethnicity')['reading score'].mean().reset_index()

# Create a bar plot using Plotly Express
fig = px.bar(race_ethnicity_avg_reading, x='race/ethnicity', y='reading score',
             title='Average Reading Score by Race/Ethnicity',
             labels={'race/ethnicity': 'Race/Ethnicity', 'reading score': 'Average Reading Score'})
fig.show()

print("Conclusion: Group E students have the highest average reading score, while Group A students have the lowest.")


# Group data by 'test preparation course' and calculate the average writing score
test_preparation_course_avg_writing = df.groupby('test preparation course')['writing score'].mean().reset_index()

# Create a bar plot using Plotly Express
fig = px.bar(test_preparation_course_avg_writing, x='test preparation course', y='writing score',
             title='Average Writing Score by Test Preparation Course',
             labels={'test preparation course': 'Test Preparation Course', 'writing score': 'Average Writing Score'})
fig.show()

print("Conclusion: Students who completed the test preparation course have a significantly higher average writing score than students who didn't.")


# Create a box plot using Seaborn to visualize the distribution of math scores by 'race/ethnicity'
plt.figure(figsize=(10, 6))
sns.boxplot(x='race/ethnicity', y='math score', data=df)
plt.title('Distribution of Math Scores by Race/Ethnicity')
plt.show()

print("Conclusion: Group E students show the highest median math score with a smaller range of scores, indicating consistency within this group. Group A students tend to have the lowest math scores with a larger range, indicating a greater spread in performance.")



# Create a scatter plot using Plotly Express to visualize the relationship between math score and reading score, colored by 'test preparation course'
fig = px.scatter(df, x='math score', y='reading score', color='test preparation course',
                 title='Relationship between Math Score and Reading Score by Test Preparation Course')
fig.show()

print("Conclusion: Students who completed the test preparation course generally have higher math and reading scores compared to those who didn't. There's a positive correlation between math and reading scores, indicating that students who perform well in math also tend to perform well in reading.")
```

### Summary:

The **correlation heatmap** emphasizes the **relationships between numerical variables** in the student performance dataset. Notably, we observe a **positive correlation** between **math score** and **reading score**, as well as between **math score** and **writing score**. This indicates that students who perform well in one subject tend to perform well in others. The heatmap is crucial for identifying **multicollinearity** and **strong correlations**, aiding in **model building** and **feature selection**.

Furthermore, students whose **parents have a master's degree** tend to achieve the **highest average math scores**, while those whose parents have **some high school education** generally score the lowest.

Analysis by student group reveals that **Group E students** attain the **highest average reading scores**, while **Group A students** exhibit the lowest scores.

In terms of test preparation, students who completed the **test preparation course** have significantly **higher average writing scores** compared to those who did not.

For math performance, **Group E students** demonstrate the **highest median math scores** with a **narrow range**, suggesting consistent performance, whereas **Group A students** have a **broader range**, indicating more variability in their scores. Additionally, students who took the **test preparation course** generally show **higher math and reading scores**. The **positive correlation between math and reading scores** further emphasizes that students excelling in math are likely to excel in reading as well.

This summary highlights the key relationships and trends in the dataset, providing insights into **student performance**, **demographic impact**, and the **effects of test preparation**.
